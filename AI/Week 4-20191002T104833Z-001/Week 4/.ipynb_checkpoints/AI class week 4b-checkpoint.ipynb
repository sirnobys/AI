```python
boston_dummy = bos_new
```

```python
from sklearn.preprocessing import scale
```

```python
bos_dummy_copy = boston_dummy.drop('CHAS',axis=1)
bos_dummy_copy.head()
```

```python
bos_scaled = scale(bos_dummy_copy)
bos_scaled
```

```python
bos_scaled_df = pd.DataFrame(bos_scaled,columns=('RM','ZN','B','DIS','M.PRICE'))
bos_scaled_df.head()
```

```python
y = bos_scaled_df['M.PRICE']
X = bos_scaled_df.drop('M.PRICE',axis=1)
```

```python
X['CHAS'] = bos_new['CHAS']
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(lr,X,y,cv=10,scoring='neg_mean_squared_error').mean())
```
