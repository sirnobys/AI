```python
#import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading dataset into DataFrame
from sklearn.datasets import load_boston
boston = load_boston()
dataset = pd.DataFrame(data=boston.data, columns=boston.feature_names)

#printing top 5
dataset.head()
```

```python
#adding the target
dataset['MEDV'] = boston.target
dataset.head()
```

```python
#inspecting dataset
dataset.shape
```

```python
#graphing data
dataset.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
dataset.describe()
```

```python
#checking for null values(NaN)
dataset.isnull().any()
```

```python
#checking which features are better correlated with the outcome
dataset_corr = dataset.corr() # without "method='pearson'" passed into corr() pearsons is calculated by default
dataset_corr
```

```python
col = dataset_corr.nlargest(6,'MEDV').index
print(col)
```

```python
correlation_map = np.corrcoef(dataset[col].values.T)
```

```python
#visualizing correlation matrix
%matplotlib inline
import seaborn as sns
sns.set(font_scale=1.0)
heatmap = sns.heatmap(correlation_map, cbar = True, annot = True, square = True, fmt='.2f',
                     yticklabels=col.values,xticklabels=col.values)
```

### Selecting features

```python
selected_variables = dataset[col]
selected_variables.head()
```

### Data Cleaning and Transformation
This dataset looks clean so we would go on to scale it.

```python
# removing CHAS from dataset

chas = selected_variables['CHAS'].copy()
other_variables = selected_variables.drop('CHAS', axis=1)
other_variables.head()
```

```python
#applying MinMax scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(other_variables)
other_variables_scaled = scaler.transform(other_variables)
```

```python
scaled_df = pd.DataFrame(data=other_variables_scaled, columns=['MEDV','RM','ZN','B','DIS'])
```

```python
#separating out the label
y_scaled = scaled_df['MEDV'].copy()
y_scaled.head()
```

```python
#adding CHAS back to features
X_scaled = scaled_df.drop('MEDV', axis=1)
X_scaled['CHAS'] = chas
X_scaled.head()
```

### Model Selection

```python
#import all algorithms needed
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
```

```python
#model selection would help us test multiple algorithms on the same dataset.
from sklearn import model_selection 
```

```python
#array of models
models = []
models.append(('LR', LinearRegression()))
models.append(('LS', Lasso()))
models.append(('GBR', GradientBoostingRegressor()))
models.append(('R', Ridge()))
models.append(('EN', ElasticNet()))
```

```python
models
```

```python
names=[]
results=[]

# Every algorithm is tested and results are
# collected and printed
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=7)
    cv_results = model_selection.cross_val_score(model, X_scaled, y_scaled, cv=kfold, scoring='neg_mean_squared_error')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
#using GridSearchCV to get the best hyperparameter combination
from sklearn.model_selection import GridSearchCV

param_grid = {
    'loss': ['ls', 'lad', 'huber', 'quantile'],
    'learning_rate': [0.1, 0.3, 0.5, 0.8, 1],
    'n_estimators': [90, 100, 200, 300, 400],
    'max_depth':[2,3,4,5,6],
    'alpha':[0.2,0.5,0.6,0.9]
}

gbr = GradientBoostingRegressor()

grid_search = GridSearchCV(
    gbr, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_scaled, y_scaled)
```
